```python
import jieba
import os

os.listdir('./res_gossiping')
jieba.load_userdict('./mydict.txt')
```

```python
sb = '' # String Builder
for fileName in os.listdir('./res_gossiping'):
    filePath = './res_gossiping/' + fileName
    try:
        with open(filePath, 'r', encoding='utf-8') as f:
            sb += f.read().split('---split---')[0]
    except:
        pass
```

```python
sb
```

```python
articleCut = jieba.cut(sb)
```

```python
articleCutList = list()
for i in articleCut:
    articleCutList.append(i)
```

```python
type(articleCut)
```

```python
wordCount = dict() # {'word1': 10, 'word2': 3}
for word in articleCutList:
    if word in wordCount:
        wordCount[word] += 1
    else:
        wordCount[word] = 1

wordCount
```

```python
# [('作者', 486), ('bbsbox', 2), ...]
wordCountList = [(k, v) for k, v in wordCount.items() if k.__len__() > 1]
wordCountList
```

```python
wordCountList.sort(key=lambda x: x[1], reverse=True)
wordCountList
```
