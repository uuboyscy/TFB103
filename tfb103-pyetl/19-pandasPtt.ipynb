```python
import pandas as pd
import glob
```

```python
glob.glob('./res_gossiping/*.txt') # Unix style
```

```python
data = list()
for filePath in glob.glob('./res_gossiping/*.txt'):
    with open(filePath, 'r', encoding='utf-8') as f:
        tmpList = f.read().split('---split---')[1].split('\n')
        data.append(tmpList[1:-1])
data
```

```python
columns = [i.split(':')[0] for i in data[0]]
columns
```

```python
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
def columnFilter(s):
    output = s.split(': ')[1]
    return output
```

```python
df['推'].apply(columnFilter)
```

```python
df
```

```python
df2 = df.copy()
# Linked List
```

```python
for c in df:
    print(c)
    df[c] = df[c].apply(columnFilter)
df
```

```python
for c in df2:
    print(c)
    df2[c] = df2[c].apply(lambda s: s.split(': ')[1])
df2
```
