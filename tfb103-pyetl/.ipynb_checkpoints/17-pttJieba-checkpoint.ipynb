```python
import jieba
import os

os.listdir('./res_gossiping')
```

```python
sb = '' # String Builder
for fileName in os.listdir('./res_gossiping'):
    filePath = './res_gossiping/' + fileName
    try:
        with open(filePath, 'r', encoding='utf-8') as f:
            sb += f.read().split('---split---')[0]
    except:
        pass
```

```python
sb
```

```python
articleCut = jieba.cut(sb)
```

```python
for i in articleCut:
    print(i)
```
